```python
import statsmodels.formula.api as smf
import pandas as pd
```

```python
df = pd.read_excel("BNPL_Survey_Unemployed.xlsx")
print(df.head())
```

```python
model_h1 = smf.logit('bnpl_used ~ finlit_score', data=df).fit()

model_h1.summary()
```

```python
model_h2 = smf.logit('bnpl_used ~ finlit_score * attitude_score', data=df).fit()
model_h2.summary()
```

```python
model_h3 = smf.logit('bnpl_used ~ finlit_score * sn_score', data=df).fit()
model_h3.summary()
```

```python
model_h4 = smf.logit('bnpl_used ~ finlit_score * pbc_score', data=df).fit()
model_h4.summary()
```
